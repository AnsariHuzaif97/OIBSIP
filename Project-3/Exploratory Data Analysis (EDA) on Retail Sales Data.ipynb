# Extending Analysis on the dataframe loaded

```python
df['Date'] = pd.to_datetime(df['Date'])#Made the Date column of type datetime

#Extracting the year out of Date and adding it as as corresponding seperate column.
df["Year"]=df["Date"].dt.year#.dt.year is used to extract the year from the datetime object.
```

```python
#Number of Transactions for each year
#=====================================
df['Year'].value_counts()
```

```python
#Sorting out the data i.e. the dataframe with respect to Date
#=============================================================
df.sort_values(by = 'Date', inplace = True)
df
```

```python
#Now we are interested to calculate total amount earned in respective years
#===========================================================================

# Group by year and sum the amount
yearly_earnings = df.groupby('Year')['Total Amount'].sum().reset_index()

yearly_earnings
```

```python
#Category wise Total Amount 
#============================

#In the year 2023
#=================
df1=df[df["Year"] == 2023]
category_total_2023=df1.groupby("Product Category")["Total Amount"].sum()
print(category_total_2023)

print("*"*40)

#In the year 2024
#=================
df2=df[df["Year"] == 2024]
category_total_2024=df2.groupby("Product Category")["Total Amount"].sum()
print(category_total_2024)
```

```python
#Quantity Wise total of products in 2023 
#=========================================
df1=df[df["Year"]==2023]
print(f"In the year 2023 : \n{"*"*40}\n{df1.groupby("Product Category")["Quantity"].sum()}")

print()
#Quantity Wise total of products in 2024
#=========================================
df2=df[df["Year"]==2024]
print(f"In the year 2024 : \n{"*"*40}\n{df2.groupby("Product Category")["Quantity"].sum()}")

print()
#Total Quantity Verification
#=============================
manualSum=0
for i in df1.groupby("Product Category")["Quantity"].sum():
    manualSum+=i
for j in df2.groupby("Product Category")["Quantity"].sum():
    manualSum+=j
print(f"Both the sum are equal: {manualSum==df["Quantity"].sum()}")    
```

```python
#Gender wise total customer
#===========================

#Year: 2023
#=============
sumMale=(df[df["Year"]==2023]["Gender"]=="Male").sum()
sumFemale=(df[df["Year"]==2023]["Gender"]=="Female").sum()
print(f"The total no of male in 2023 is : {sumMale} and female is : {sumFemale}")

#Justification
#==============
print(f"The sum are equal : {((df[df["Year"]==2023]["Gender"]).count())==sumMale+sumFemale}")
print()

#Year: 2024
#=============
sumMale=(df[df["Year"]==2024]["Gender"]=="Male").sum()
sumFemale=(df[df["Year"]==2024]["Gender"]=="Female").sum()
print(f"The total no of male in 2024 is : {sumMale} and female is : {sumFemale}")

#Justification
#==============
print(f"The sum are equal : {((df[df["Year"]==2024]["Gender"]).count())==sumMale+sumFemale}")
print()
```

# Total sales in every month of the year : 2023

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))

dataFrame=((df[df["Year"]==2023]).groupby(by="Months")["Total Amount"].sum())
month_order = ['January', 'February', 'March', 'April', 'May', 'June','July', 'August', 'September', 'October', 'November', 'December']
dataFrame.index=month_order

print(plt.plot(dataFrame.index, dataFrame.values, marker='o', color='blue'))
plt.title("Total Sales in every month of the year : 2023")
plt.xlabel("Months")
plt.ylabel("Total Amount")
plt.grid()
```

# Total sales in every month of the year : 2024

```python
plt.figure(figsize=(10,6))

dataFrame=((df[df["Year"]==2024]).groupby(by="Months")["Total Amount"].sum())
month_order = ['January', 'February', 'March', 'April', 'May', 'June','July', 'August', 'September', 'October', 'November', 'December']
final_index=dict(list(zip(month_order,[1]*12)))
for i in final_index:
    if i in dataFrame.index:
        pass
    else:
        final_index[i]=0
        
print(plt.plot([j for j in final_index if final_index[j]==1], dataFrame.values, marker='o', color='Green'))
plt.title("Total Sales in every month of the year : 2024")
plt.xlabel("Months")
plt.ylabel("Total Amount")
plt.grid()
#Since in 2024 we have only month due to which we get a dot.
```

# Total Sales category wise in the the year 2023

# Determining which gender has made largest purchase

```python
df1=df.groupby(by="Gender")["Total Amount"].sum()

plt.figure(figsize=(10,6))
plt.pie(df1.values, labels=df1.index, autopct='%2f%%')
plt.title("Gender-wise Maximun Sale")

if df1["Female"]>df1["Male"]:
    print("Femles has made largest purchase")
else:
    print("Male has made largest purchase")

print("*"*40)
print()
```
